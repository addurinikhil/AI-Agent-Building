```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
loader = PyPDFLoader("/home/nineleaps/Documents/Ai_Agent_Build_Training/Day_1_Study_Resource .pdf")
documents = loader.load()

print(f"Pages Loaded: {len(documents)}")
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=400,
    chunk_overlap=40
)

chunks = splitter.split_documents(documents)

print(f"Total Chunks: {len(chunks)}")
```

```python
chunk_texts = [chunk.page_content for chunk in chunks]
```

```python
vectorizer = CountVectorizer()

vector_db = vectorizer.fit_transform(chunk_texts)

print("Vector Shape:", vector_db.shape)
```

```python
question = input("Enter your question: ")
```

```python
query_vector = vectorizer.transform([question])
```

```python
scores = cosine_similarity(query_vector, vector_db)

top_chunks = scores[0].argsort()[-2:][::-1]

print("Most Relevant Chunks:", top_chunks)
```

```python
retrieved_context = []

for idx in top_chunks:
    retrieved_context.append(chunk_texts[idx])

context = "\n".join(retrieved_context)
```

```python
print("\nAnswer Based On Retrieved Context:\n")
print(context)
```
