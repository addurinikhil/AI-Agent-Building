```python
#Load PDF
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("rag_notes.pdf")
docs = loader.load()

print(docs[0].page_content)
```

```python
# split documents
```

```python
!pip install -q langchain-text-splitters
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_documents(docs)

print(f"Number of chunks: {len(chunks)}")
print(chunks[0].page_content)
```

```python
# Create embeddings
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
# Create vector database
```

```python
from langchain_community.vectorstores import Chroma

vectordb = Chroma.from_documents(
    chunks,
    embeddings
)
```

```python
# create retriever
```

```python
retriever = vectordb.as_retriever(
    search_kwargs={"k":3}
)
```

```python
retriever.invoke("What is RAG?")
```

```python
from datetime import datetime
```

```python
# add conversational memory
```

```python
chat_history = []
def add_to_history(role, content):
    chat_history.append({
        "role": role,
        "content": content
    })
```

```python
#create history formatter
```

```python
def get_chat_history():

    history = ""

    for msg in chat_history:
        history += f"{msg['role']}: {msg['content']}\n"

    return history
```

```python
# setup LLM
```

```python
!pip install langchain-google-genai

from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key="GEMINI_API_KEY"
)
```

```python
#Create History-Aware Question Rewriter
```

```python
def rewrite_question(question):

    history = get_chat_history()

    prompt = f"""
You are a question rewriting assistant.

Conversation History:
{history}

Current Question:
{question}

Rewrite the current question into a standalone question.

Return only the rewritten question.
"""

    response = llm.invoke(prompt)

    return response.content
```

```python
rewrite_question("Explain it more")
```

```python
#Create Retrieval Function
```

```python
def retrieve_context(question):

    docs = retriever.invoke(question)

    context = "\n\n".join(
        [doc.page_content for doc in docs]
    )

    return context
```

```python
#Create RAG Answer Function
```

```python
def answer_from_rag(question):

    context = retrieve_context(question)

    prompt = f"""
Answer using ONLY the provided context.

Context:
{context}

Question:
{question}

Answer:
"""

    response = llm.invoke(prompt)

    return response.content
```

```python
#Tool Decision Function
```

```python
def tool_required(question):

    keywords = [
        "time",
        "current time",
        "date",
        "today"
    ]

    q = question.lower()

    for word in keywords:

        if word in q:
            return True

    return False
```

```python
#Execute Tool
```

```python
def execute_tool(question):

    if tool_required(question):

        current_time = get_current_time()

        return f"Current Time: {current_time}"

    return "No suitable tool found."
```

```python
#Relevance Check
```

```python
def context_exists(question):

    docs = retriever.invoke(question)

    context = " ".join(
        [doc.page_content for doc in docs]
    )

    return len(context.strip()) > 50
```

```python
#tool required
def tool_required(question):

    q = question.lower()

    keywords = [
        "time",
        "current time",
        "date",
        "today"
    ]

    return any(word in q for word in keywords)
```

```python
#Routing Logic
```

```python
def route_question(question):

    # Step 1: Rewrite follow-up question
    standalone_question = rewrite_question(question)

    print("\nStandalone Question:")
    print(standalone_question)

    # Debug Information
    print("\nTool Required:", tool_required(standalone_question))
    print("Context Exists:", context_exists(standalone_question))

    # Step 2: Tool Routing
    if tool_required(standalone_question):

        print("\nUsing Tool Call...")

        answer = execute_tool(
            standalone_question
        )

        source = "TOOL"

    # Step 3: RAG Routing
    elif context_exists(standalone_question):

        print("\nUsing RAG Retrieval...")

        answer = answer_from_rag(
            standalone_question
        )

        source = "RAG"

    # Step 4: Fallback
    else:

        answer = (
            "I couldn't find relevant information "
            "in the knowledge base and no suitable tool "
            "was available."
        )

        source = "NONE"

    return answer, source
```

```python
#Full Chat Loop
```

```python
while True:

    user_input = input("\nYou: ")

    if user_input.lower() == "exit":
        break

    add_to_history(
        "user",
        user_input
    )

    answer, source = route_question(
        user_input
    )

    print(f"\nAssistant ({source}):")
    print(answer)

    add_to_history(
        "assistant",
        answer
    )
```
